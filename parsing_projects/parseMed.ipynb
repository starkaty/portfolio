```python
from pprintpp import pprint as pp
import pandas as pd
import numpy as np
import requests
from lxml import html
```

```python
url1 = 'https://www.medica-tradefair.com/kati-cgi/kati/vis/custom/ext2/download.cgi?ticket=g_u_e_s_t&lang=2&fair=medcom2022&dl=html'
```

```python
url = url1
headers = {'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:109.0) Gecko/20100101 Firefox/109.0'}
res = requests.get(url, headers = headers)
page = html.fromstring(res.text)
company_name = page.xpath('//tr/td[1]/text()')
```

```python
pp(company_name)
```

```python
#создаем датафрейм pandas
medica_df = pd.DataFrame({'company_name': company_name[i]} for i in range(len(company_name)))
```

```python
medica_df.head()
```

```python
medica_df.info()
```

```python
medica_df['company_name'] = medica_df['company_name'].apply(lambda x: x.upper())
```

```python
medica_df['ex'] = 'medica'
```

```python
medica_df.head()
```

```python
medica_df.drop(labels=60, axis=0, inplace=True)
```
