```python
import requests
import json
from pprintpp import pprint as pp
```

```python
url = 'https://api.github.com' #ссылка на ресурс
```

```python
user = 'starkaty' #имя пользователя
```

```python
res = requests.get(f'{url}/users/{user}/repos') #создаем объект responce
```

```python
pp(res.json())#посмотрим содержимое ответа сервера в формате json
```

```python
#запишем файл json
with open('repos.json', 'w') as f:
    json.dump(res.json(), f)
```

```python
#выведем список репозиториев указанного пользователя
for el in res.json():
    print (el['name'])
```
