# TensorFlow. Dense

```python
import tensorflow as tf
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston

from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import RobustScaler

import matplotlib.pyplot as plt

from tensorflow.keras.optimizers import Adam, SGD


from sklearn.metrics import r2_score

tf.random.set_seed(1)
```

```python
boston = load_boston()
X = boston.data
y = boston.target
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test =train_test_split(X, y, test_size = 0.25, random_state = 100)
```

```python
scaler = RobustScaler()
scaler = scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
```

## Создадим 10 различных моделей полносвязной нейросети с различныими параметрами и сравним их по метрике R2. Так как мы решаем задачу регрессии, эта метрика представляется оптимальной. 

## Также в таблицу будем записывать метрики mse и mae.

# Модель 1

```python
model_num = 1

params = '13 input, 1 hidden layer (10), act = relu, epochs = 100, batch size = 10, opt = rmsprop, loss = mse'

inputs = Input(shape=(13))
hidden_layer_1 = Dense(10, activation='relu')(inputs)
outputs = Dense(1)(hidden_layer_1)

model = Model(inputs, outputs)

model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])

history = model.fit(x_train, 
                    y_train, 
                    epochs=100,
                    batch_size = 10,
                    validation_split=0.1, 
                    verbose=2)


y_test_pred = model.predict(x_test)
test_mse, test_mae = model.evaluate(x_test, y_test, verbose=2)

test_r2 = r2_score(y_test, y_test_pred)

result_list.append([model_num, params, test_mse, test_mae, test_r2])
print 
(f'model_num: {model_num}, params: ({params}), test_mse: {round(test_mse, 2)}, test_mae: {round(test_mae,2)}, test_r2: {round(test_r2,4)}')
```
