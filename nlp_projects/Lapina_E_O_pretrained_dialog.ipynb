# Загрузка предобученной модели для поддержания разговора (sber, на gpt2)

```python
def talking(text, max_len):
    inp = text
    encoded = tokenizer.encode(inp, return_tensors="pt")


    result = ru_gpt.generate(encoded,
        max_length=max_len,
        early_stopping=True,
        no_repeat_ngram_size=5)
    result= tokenizer.decode(result.cpu().flatten().numpy(), skip_special_tokens=True)
    try:        
        return result
    except:
        return 'Давай поговорим о чем-нибудь другом'
```

# Болталка 

```python
print(talking('ты хочешь есть?'), 1000)
```

# Погода в городе

```python
print(talking('Какая завтра будет погода?'), 1000)
```

# Праздники

```python
print(talking('Как отмечать Новый год?'), 1000)
```

# Фильмы

```python
print(talking('Посоветуй хороший фильм?'), 1000)
```

# Покупка товара
