```python
texts = []
images = []
titles = []
links = []
```

```python
service = Service('chromedriver.exe')
driver = webdriver.Chrome(service=service)
driver.maximize_window()
n=10000
for link in df_links['link'][10000:]:
    print(link, 'ссылка', n)
    n +=1
    if n % 50 == 0:
        service = Service('chromedriver.exe')
        driver = webdriver.Chrome(service=service)
        driver.maximize_window()
    try:
        driver.get(link)
        title = driver.find_elements(By.XPATH, '//div/h1[@class="BookCard-module__book__mainInfo__title_2zz4M"]')[0].text
        text = driver.find_elements(By.XPATH, '//div[@class="BookCard-module__book__annotation_2ZIhf"]/p')[0].text
        image = driver.find_elements(By.XPATH, '//img[@loading="eager"]')[0].get_attribute('src')
        texts.append(text)
        images.append(image)
        titles.append(title)
        links.append(link)
    except:
        continue
    
df =  pd.DataFrame(titles, columns = ['title'])
df['texts'] = texts
df['images'] = images
df['link'] = links
df.to_csv('books_list_litres1.csv')
```
